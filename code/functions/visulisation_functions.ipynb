```python
def y_over_time(y):
    #Plot values of Y over query number (index is a proxy)
    plt.figure(figsize=(12, 5))
    plt.plot(range(len(y)), y, marker='o', linestyle='-', color='blue')
    plt.title("Output Values by Submission")
    plt.xlabel("Submission Number")
    plt.ylabel("Output Values")
    plt.grid(True, which='both', linestyle='--', alpha=0.7)
    
    plt.show()
```
